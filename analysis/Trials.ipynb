```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
from matplotlib.colors import ListedColormap
```

```python
# Example DataFrame with p attributes and target y
n = 100  # number of data points
p = 4    # number of attributes

# Generate random data for demonstration purposes
np.random.seed(0)
data = np.random.randn(n, p)
y = np.random.choice([-1, 1], size=n)

# Create a DataFrame
columns = [f'attr_{i+1}' for i in range(p)]
df = pd.DataFrame(data, columns=columns)
df['y'] = y
```

```python
def calculate_point_sizes(df, x_col, y_col):
    # Group by the two attributes and count occurrences
    group_sizes = df.groupby([x_col, y_col]).size().reset_index(name='size')
    
    # Merge the group sizes back into the original dataframe
    df = pd.merge(df, group_sizes, on=[x_col, y_col], how='left')
    
    return df['size']
```

```python
def scatter_matrix(df):
    p = df.shape[1] - 1  # Number of attributes, excluding target variable
    attributes = df.columns[:-1]
    
    cmap = ListedColormap(['green', 'red'])  # Custom colormap for y=-1 (green) and y=1 (red)
    
    fig, axes = plt.subplots(nrows=p, ncols=p, figsize=(15, 15))
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for i in range(p):
        for j in range(p):
            ax = axes[i, j]
            
            if i == j:
                # Diagonal plots can be histograms of the attribute
                ax.hist(df[attributes[i]], bins=30, color='gray', alpha=0.5)
                ax.set_xlabel(attributes[i])
                ax.set_ylabel(attributes[i])
            else:
                sizes = calculate_point_sizes(df, attributes[j], attributes[i])
                
                # Scatter plot
                scatter = ax.scatter(df[attributes[j]], df[attributes[i]], 
                                     c=df['y'], cmap=cmap, s=sizes*10, alpha=0.6, edgecolors='w')
                
                ax.set_xlabel(attributes[j])
                ax.set_ylabel(attributes[i])

    plt.show()

# Call the function to plot the scatter matrix
scatter_matrix(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm

# Step 1: Generate some example data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)+1  # Values used for coloring

# Step 2: Define the colormap and normalization
cmap = plt.get_cmap('RdYlGn')  # Red for positive, Green for negative
norm = TwoSlopeNorm(vmin=-np.max(np.abs(z)), vcenter=0, vmax=np.max(np.abs(z)))

# Step 3: Create the scatter plot
plt.scatter(x, y, c=z, cmap=cmap, norm=norm)

# Step 4: Add a colorbar
plt.colorbar(label='Value')

# Step 5: Add labels and show the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Diverging Colormap (0 = White)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
values = np.random.rand(50) * 10  # Values corresponding to colors

# Create a scatter plot with unfilled markers
plt.scatter(x, y, c=values, cmap='viridis', facecolors='none', edgecolors=plt.cm.viridis(values / 10))

# Add a color bar
plt.colorbar(label='Value')

# Add labels and show the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Unfilled Markers Colored by Values')
plt.show()
```
